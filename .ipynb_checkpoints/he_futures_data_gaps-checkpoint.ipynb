```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
hog_vol_df = pd.read_csv("../data/he_futures_data.csv", names=["Index", "Product", "Contract", 
                                                                  "Price", "Date", "???"])

# slice by contract right off the bat
exp_months = hog_vol_df["Contract"].unique()

contract_vol_dict = {}

for month in exp_months:
    mon_df = hog_vol_df.loc[hog_vol_df["Contract"] == month]
    mon_df.set_index("Date", inplace=True)
    
    mon_df.index = pd.to_datetime(mon_df.index)
    
    contract_vol_dict[month] = mon_df
    
contract_vol_dict["JUL18"]
```

```python
# large gaps

jul_df = contract_vol_dict["JUL19"].copy()

jul_df.loc[jul_df.index > dt.datetime(year=2018, month=11, day=5, hour=15, minute=20)].head(20)
```

```python
# more gaps

jul_df.loc[jul_df.index > dt.datetime(year=2018, month=8, day=17, hour=14, minute=30)].head(20)
```

```python
# weird gap right when the JUL19 contract starts trading

jul_df.loc[jul_df.index > dt.datetime(year=2018, month=8, day=1, hour=18, minute=0)].head(20)
```

```python
# more gaps -  this is right at the start of the dataset

dec_df = contract_vol_dict["DEC18"].copy()

dec_df.loc[dec_df.index > dt.datetime(year=2018, month=2, day=2, hour=13, minute=30)].head(10)
```

```python
# smaller gaps

dec_df = contract_vol_dict["DEC18"].copy()

dec_df.loc[dec_df.index > dt.datetime(year=2018, month=6, day=6, hour=15, minute=25)].head(20)
```

```python
# data goes to daily frequency after 11/16/18
# april 2019 contract as example
for contract in contract_vol_dict.keys():

    con_df = contract_vol_dict[contract].copy()
    print(con_df.loc[con_df.index > dt.datetime(year=2018, month=11, day=16, hour=19, minute=4)].head(20))
```

```python
# show how much intra day data there is after 11/16/18
for contract in contract_vol_dict.keys():
    
    con_df = contract_vol_dict[contract].copy()
    con_df["Day"] = con_df.index.date
    con_df = con_df.loc[con_df["Day"] > dt.date(year=2018, month=11, day=16)]
    # skip contracts that expire prior to 11/16/18
    if len(con_df) == 0:
        continue
    # drop all observations that are the only observation for the given day
    con_df = con_df.loc[(con_df["Day"] == con_df["Day"].shift(1)) | (con_df["Day"] == con_df["Day"].shift(-1))]
    # count the remaining rows
    print(f"contract {contract} has {len(con_df)} rows of data at greater than daily frequency past 11/16/18")
```
