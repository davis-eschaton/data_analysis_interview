```python
import pandas as pd
import numpy as np
import datetime as dt

%load_ext autoreload
%autoreload 2

pd.options.display.max_rows = 2000
pd.options.display.max_columns = 2000
```

```python
# slicing by expiration month
# pass in a number of cols of a product

FC = pd.read_csv("../data/FC1.csv")
exp_months = FC["Exp_Month"].unique()

for m in range(2,8):
    m_read = pd.read_csv(f"../data/FC{m}.csv")
    FC = FC.append(m_read, ignore_index=True)

#print(FC)

# store the new dataframes in a dictionary but can easily export all of these to csvs. Would just be a lot
gf_months = {}

for month in exp_months:
    mon_df = FC.loc[FC["Exp_Month"] == month]
    mon_df.set_index("Date", inplace=True)
    gf_months[month] = mon_df
```

```python
# slicing by expiration month
# pass in a number of cols of a product

LC = pd.read_csv("../data/LC1.csv")
exp_months = LC["Exp_Month"].unique()

for m in range(2,8):
    m_read = pd.read_csv(f"../data/LC{m}.csv")
    LC = LC.append(m_read, ignore_index=True)

#print(LC)

# store the new dataframes in a dictionary but can easily export all of these to csvs. Would just be a lot
lc_months = {}

for month in exp_months:
    mon_df = LC.loc[LC["Exp_Month"] == month]
    mon_df.set_index("Date", inplace=True)
    lc_months[month] = mon_df
```

```python
gf_aug = [gf_months[month] for month in exp_months if 'AUG' in month]
lc_feb = [lc_months[month] for month in exp_months if 'FEB' in month]
```

```python
gf = pd.concat(gf_aug)
lc = pd.concat(lc_feb)
```

```python
gf.tail()
```

```python
df = gf.merge(lc, how='outer', left_index=True, right_index=True, suffixes=["_FC", "_LC"])
```

```python
zc = pd.read_csv("../data/C.csv").set_index('Date')
zc.index = pd.to_datetime(zc.index)
```

## year over year trends in the crush/behavior of LE after

```python
import matplotlib.pyplot as plt
import datetime

years = {}
for i in range(21):
    yr_df = df[(pd.to_datetime(df.index) > datetime.datetime(2000 + i, 5, 1)) & (pd.to_datetime(df.index) < datetime.datetime(2000 + i, 12, 1))]
    yr_df['Feeder Profit/CWT'] = (6 * 400 * yr_df['Settle_LC'].values - 2 * 50 * yr_df['Settle'].values - 3 * 500 * yr_df['Settle_FC'].values) / 1500
    years[2000 + i] = yr_df
```

```python
for i, (year, yr_df) in enumerate(years.items()):
    print(f"YEAR: {year}")
    print(f"EXPIRATION: {yr_df.index[-1]}")
    fig, ax1 = plt.subplots()

    ax1.plot(yr_df[(pd.to_datetime(yr_df.index) > datetime.datetime(2000 + i, 5, 1)) & (pd.to_datetime(yr_df.index) < datetime.datetime(2000 + i, 9, 1))]['Feeder Profit/CWT'], label='Feeder Profit/CWT', color='blue')
    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
    ax2.plot(yr_df[(pd.to_datetime(yr_df.index) > datetime.datetime(2000 + i, 5, 1)) & (pd.to_datetime(yr_df.index) < datetime.datetime(2000 + i, 12, 1))]['Settle_LC'], label='Cattle', color='orange')
    
    # plt.xticks(np.arange(0, len(yr_df['Settle']) + 1, 20))
    plt.xticks(rotation=45)

    plt.legend()
    plt.show()
```

### Doe the crush have directionality?

#### Yearly

```python
for exp, ydf in years.items():
    ydf = ydf[pd.to_datetime(ydf.index) < datetime.datetime(exp, 9, 1)]
    ydf['Crush'] = (2400 * ydf['Settle_LC'] - 100 * ydf['Settle'] - 1500 * ydf['Settle_FC']) / 1500
    print(f"YEAR: {exp}")
    plt.scatter(ydf['Settle_FC'] + ydf['Settle_LC'], ydf['Crush'])
    plt.show()
```

#### Aggregate

```python
all_df = pd.concat(years.values())
all_df['Crush'] = (2400 * all_df['Settle_LC'] - 100 * all_df['Settle'] - 1500 * all_df['Settle_FC']) / 1500
plt.scatter(all_df['Settle_FC'] + all_df['Settle_LC'], all_df['Crush'])


plt.figure(figsize=(15, 5))
plt.show()
```

```python
all_df.dropna(inplace=True)
x, y = all_df['Settle_FC'] + all_df['Settle_LC'], all_df['Crush']

correlation_matrix = np.corrcoef(x, y)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
```
