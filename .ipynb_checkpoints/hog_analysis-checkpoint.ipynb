```python
import pandas as pd
import numpy as np
import datetime as dt
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Read in data

First need to read the hog index data, construct the 1-day and 2-day indices, and store it all in a dataframe.

```python
# read in hog index data and make that a new dataframe
raw_hog_df = pd.read_csv("../data/HogSheet.csv", header=0, names=["Reported_For_Date", "Purch_Type", "Head_Ct", 
                                                                  "Avg_Net_Price", "Avg_Carcass_Wt"])
raw_hog_df = raw_hog_df.apply(pd.to_numeric, errors="ignore")
raw_hog_df['Head_Ct'] = raw_hog_df['Head_Ct'].str.replace(',', '').astype(float)

# new dataframe that will be the final
hog_df = raw_hog_df.loc[raw_hog_df["Purch_Type"] == "Prod. Sold Negotiated"]
hog_df["Tot_Wt"] = hog_df["Head_Ct"]*hog_df["Avg_Carcass_Wt"]
hog_df["Tot_Val"] = hog_df["Tot_Wt"]*hog_df["Avg_Net_Price"]

# discard the rest of the data except for the second kind of purchase type needed
# bummer to repeat these same lines of code
raw_hog_df = raw_hog_df.loc[raw_hog_df["Purch_Type"] == "Prod. Sold Swine or Pork Market Formula"]
raw_hog_df["Tot_Wt"] = raw_hog_df["Head_Ct"]*raw_hog_df["Avg_Carcass_Wt"]
raw_hog_df["Tot_Val"] = raw_hog_df["Tot_Wt"]*raw_hog_df["Avg_Net_Price"]

hog_df = hog_df.join(raw_hog_df.set_index("Reported_For_Date"), on="Reported_For_Date", rsuffix="_2")

# could do this by concat and summing dataframes but i think would require dropping purch_type col from both
hog_df["Agg_Tot_Wt"] = hog_df["Tot_Wt"] + hog_df["Tot_Wt_2"]
hog_df["Agg_Tot_Val"] = hog_df["Tot_Val"] + hog_df["Tot_Val_2"]

hog_df["1_Day_Indx"] = hog_df["Agg_Tot_Val"]/hog_df["Agg_Tot_Wt"]
hog_df["2_Day_Indx"] = (hog_df["Agg_Tot_Val"]+hog_df["Agg_Tot_Val"].shift(-1))/(hog_df["Agg_Tot_Wt"]+\
                                                                                hog_df["Agg_Tot_Wt"].shift(-1))

# could drop intermediate columns but don't need to
hog_df.set_index("Reported_For_Date", drop=True, inplace=True)
hog_df.index = pd.to_datetime(hog_df.index)

hog_df

# don't think i need to move dates ahead by one? Original hog sheet appears to do so
```

### 1A. Read in futures data

Now want to read in the futures data created from the high frequency sheet. Everything should be exported to csv from the hog_vol_high_freq sheet. The csv names are just the contract names but not sure of the best way to read in all of these... For now just going to do it with a list but will need to reorganize notebooks/figure something else out in the future. Could semi-easily loop through expiration months and a range of years if we knew that we had data for a nice chunk of time like that but the months we have data for from the he_futures_data sheet are kinda spotty.

```python
contract_dict = {}

# a lot of the data from the high frequency set is pretty poor
for contract in ["APR18", "MAY18", "JUN18", "JUL18", "AUG18", "OCT18"]:
    
    contract_df = pd.read_csv(f"../data/{contract}.csv")
    
    contract_df.set_index("Date", inplace=True)
    contract_df.index = pd.to_datetime(contract_df.index)
    
    # only keep if the date 20 days before future expiry is in the range of the index data and if the future expires before the index data ends
    if (contract_df.index[-21] >= hog_df.index[-1]) and (contract_df.index[-1] <= hog_df.index[0]):
        
        # reverse dataframe to keep semi-compatible with the rest of this notebook
        contract_df = contract_df[::-1]
        # theoretically could also rename price and abs_change cols to settle and change be consistent
        
        contract_dict[contract] = contract_df
        
        
        
contract_dict["MAY18"]
```

### 1B. Compile dataframe for regressions

Since we want to do a number of regressions at various lengths of time from expiration, we need to put all of that information in a dataframe. Unfortunately, the high frequency data only covers 6 contracts, so the dataframe isn't very long. Also, we're using the 2 day index for everything.

```python
# each row corresponding to a contract with cols for index difference, forward vol, etc.

reg_df = pd.DataFrame()
reg_df["Contract"] = contract_dict.keys()


for t in [20, 10, 5, 2]:    
    # realized vol over the last t days of the contract - target
    reg_df[f"Fut_Vol_{t}"] = reg_df["Contract"].apply(lambda x: contract_dict[x].loc[contract_dict[x].index[0], f"Vol_{t}"])
for t in [20, 10, 5, 2]:
    # should be using t or t-1? is the last day 1 or 0 days to expiry? currently treating the last day as 0 days to expiry
    # think we need to use t because the t-day volatility is the volatility of the final two days of log returns and we should be predicting that based on a statistic from t+1
    # days away from the expiration of the contract because the t-day log returns include information from t+1 days. So the index is t because that's actually the t+1th row 
    # of the dataframe
    
    # absolute difference in future & 2 day rolling index t days from expiry
    reg_df[f"Diff_{t}"] = reg_df["Contract"].apply(lambda x: abs(contract_dict[x].loc[contract_dict[x].index[t],"Price"] - hog_df.loc[contract_dict[x].index[t],"2_Day_Indx"]))
    # reg_df[f"Diff1_{t}"] = reg_df["Contract"].apply(lambda x: abs(contract_dict[x].loc[contract_dict[x].index[t],"Settle"] - hog_df.loc[contract_dict[x].index[t],"1_Day_Indx"]))
for t in [20, 10, 5, 2]:    
    # absolute change in price over the last t days of the contract
    reg_df[f"Chg_{t}"] = reg_df["Contract"].apply(lambda x: contract_dict[x].loc[contract_dict[x].index[0],f"Abs_Change_{t}"])
for t in [20, 10, 5, 2]:
    # s-day volatility t days out from expiry. 
    # use index t for same reason as above: t day fut vol uses last t log returns, which includes settle price on t+1th day, so use volon t+1th day to predict.
    for s in [20, 10, 5, 2]:
        reg_df[f"Vol_{t}_{s}day"] = reg_df["Contract"].apply(lambda x: contract_dict[x].loc[contract_dict[x].index[t],f"Vol_{s}"])
for t in [20, 10, 5, 2]:
    # future volatility divided by current level of volatility (past s days)
    for s in [20, 10, 5, 2]:
        reg_df[f"Fut_Vol_{t}_Rel{s}"] = reg_df[f"Fut_Vol_{t}"]/reg_df[f"Vol_{t}_{s}day"]


reg_df
    
    
```

## 2. Regressions

Essentially we're just trying to explain future volatility at various distances from expiration, so we run a bunch of different regressions using the columns we created in the previous cell. We print out key statistics from each regression.

```python
# would be easy to change this to add more predictors or more statistics
reg_results = pd.DataFrame(columns = ["y", "X1", "X2", "Rsq", "pvalX1", "pvalX2"])

# could set lists that t, s, and d iterate through up here
# also could have defined a function that runs the regression and appends results...

# doing some regressions, mostly just variations of using the spread to predict future vol or change in vol

for t in [20, 10, 5, 2]:
    for s in [20, 10, 5, 2]:
#         for d in [1, 2]:
        y = f"Fut_Vol_{t}_Rel{s}"
        Xcol = [f"Diff_{t}"]
        X = sm.add_constant(reg_df[Xcol])
        regr = sm.OLS(reg_df[y],X).fit() # .fit(x, y)
        params = regr.params
        reg_results = reg_results.append({"y": y,
                                          "X1": Xcol[0],
                                          "Rsq": regr.rsquared,
                                          "pvalX1": regr.pvalues[Xcol[0]],
                                          "slope": params[1],
                                          "intercept": params[0]}, ignore_index=True)

for t in [20, 10, 5, 2]:  
#     for d in [1, 2]:
    y = f"Fut_Vol_{t}"
    Xcol = [f"Diff_{t}"]
    X = sm.add_constant(reg_df[Xcol])
    regr = sm.OLS(reg_df[y],X).fit() # .fit(x, y)
    params = regr.params
    reg_results = reg_results.append({"y": y,
                                      "X1": Xcol[0],
                                      "Rsq": regr.rsquared,
                                      "pvalX1": regr.pvalues[Xcol[0]],
                                          "slope": params[1],
                                          "intercept": params[0]}, ignore_index=True)

for t in [20, 10, 5, 2]:
    for s in [20, 10, 5, 2]:
#         for d in [1, 2]:
        y = f"Fut_Vol_{t}"
        Xcol = [f"Diff_{t}", f"Vol_{t}_{s}day"]
        X = sm.add_constant(reg_df[Xcol])
        regr = sm.OLS(reg_df[y],X).fit() # .fit(x, y)
        params = regr.params
        # print(params)
        reg_results = reg_results.append({"y": y,
                                          "X1": Xcol[0],
                                          "X2": Xcol[1],
                                          "Rsq": regr.rsquared,
                                          "pvalX1": regr.pvalues[Xcol[0]],
                                          "pvalX2": regr.pvalues[Xcol[1]],
                                          "slope2": params[2],
                                          "slope": params[1],
                                          "intercept": params[0]}, ignore_index=True)

for t in [20, 10, 5, 2]: 
#     for d in [1, 2]:
    y = f"Chg_{t}"
    Xcol = [f"Diff_{t}"]
    X = sm.add_constant(reg_df[Xcol])
    regr = sm.OLS(reg_df[y],X).fit() # .fit(x, y)
    params = regr.params
    reg_results = reg_results.append({"y": y,
                                      "X1": Xcol[0],
                                      "Rsq": regr.rsquared,
                                      "pvalX1": regr.pvalues[Xcol[0]],
                                          "slope": params[1],
                                          "intercept": params[0]}, ignore_index=True)

    
reg_results

# the warnings here are because there are a handful of instances where vol for the last two days is zero, i.e. the log return for the last two days is exactly the same
```

Conclusions: The results here look pretty bad in general. Regressing the spread on the future change over the last t days doesn't appear to work at all even though the parameters are consistently positive. Adding the level of vol at the time of the difference to the regression appears to make it a lot better, because clearly it is correlated with the future level of volatility, but the diff variable is insignificant. The remaining regressions do not show much evidence of a relationship between the spread and future volatility. I'll try bucketing the magnitude of the diff variable and doing regressions within the different buckets next to see if there's a relationship when the spread is higher/lower.

### 2A. Regressions bucketed by magnitude of index-future spread

First check to see what the distribution of the spread looks like.

```python
# regression bucketed by levels of diff column

# want to see if there's any clear-cut levels of diffs
plt.plot(reg_df.index, reg_df["Diff_20"], "o")
plt.plot(reg_df.index, reg_df["Diff_10"], "o")
plt.plot(reg_df.index, reg_df["Diff_5"], "o")
plt.plot(reg_df.index, reg_df["Diff_2"], "o")
```

Hard to tell much from that plot. In the future would be better to just find the mean & standard deviation and use those to create buckets.

Here we use pd.cut to create 3 bins over equal ranges.

```python
# this is what pd.cut output looks like
pd.cut(reg_df["Diff_20"], bins=3, labels=False, retbins=True)
```

Create bins for different distances t from expiration.

```python
# add columns with bucket indicators
# might want to try different bin sizes across diff ts
binsize = 2
for t in [20, 10, 5, 2]: 
    cut = pd.cut(reg_df[f"Diff_{t}"], bins=binsize, labels=False)
    print(cut)
    # check how many values are in each bucket - higher buckets have much fewer
    print(cut.value_counts())
    reg_df[f"Diff_{t}_bin{binsize}"] = cut
# low bin number means low diff value

reg_df
```

Use these bins to do the same regressions as before.

```python
# try regressions, adding results to new reg_results dataframe or previous dataframe
reg_results = pd.DataFrame(columns = ["y", "X1", "X2", "Rsq", "pvalX1", "pvalX2"])

# set size of bins to use
binsize = 2

# pretty much the same regressions as before, just using the bucketed data
for t in [20, 10, 5, 2]:  
    for b in range(binsize):
        # check to make sure we have an adequate number of values to regress with
        if len(reg_df.loc[reg_df[f"Diff_{t}_bin{binsize}"] == b]) < 3:
            continue
        y = f"Fut_Vol_{t}"
        Xcol = [f"Diff_{t}"]
        Xdata = reg_df.loc[reg_df[f"Diff_{t}_bin{binsize}"] == b]
        X = sm.add_constant(Xdata[Xcol])
        regr = sm.OLS(Xdata[y],X).fit() # .fit(x, y)
        params = regr.params
        reg_results = reg_results.append({"y": y,
                                          "X1": f"{Xcol[0]}_bin{b}",
                                          "Rsq": regr.rsquared,
                                          "pvalX1": regr.pvalues[Xcol[0]],
                                              "slope": params[1],
                                              "intercept": params[0]}, ignore_index=True)
        
for t in [20, 10, 5, 2]:
    for s in [20, 10, 5, 2]:
        for b in range(binsize):
            # check to make sure we have an adequate number of values to regress with
            if len(reg_df.loc[reg_df[f"Diff_{t}_bin{binsize}"] == b]) < 3:
                continue
            y = f"Fut_Vol_{t}_Rel{s}"
            Xcol = [f"Diff_{t}"]
            Xdata = reg_df.loc[reg_df[f"Diff_{t}_bin{binsize}"] == b]
            X = sm.add_constant(Xdata[Xcol])
            regr = sm.OLS(Xdata[y],X).fit() # .fit(x, y)
            params = regr.params
            reg_results = reg_results.append({"y": y,
                                              "X1": f"{Xcol[0]}_bin{b}",
                                              "Rsq": regr.rsquared,
                                              "pvalX1": regr.pvalues[Xcol[0]],
                                              "slope": params[1],
                                              "intercept": params[0]}, ignore_index=True)

reg_results
```

The most promising results seem to be coming from regressions of the higher spread bins on future volatility, especially at t=10 and t=5 days from expiry (rows 3 and 5) as they have high r-squared values and the predictors have significant p values and parameters seem reasonable. However, the fact that these results don't hold as much for 20 and 2 days out (rows 1 and 7) casts doubt on a tradeable relationship. Also, I'm beginning to think that it makes more sense to regress on the future volatility relative to the current level of volatility as it's more intuitive to me to try to predict how volatility is going to change rather than the level of volatility. These regressions look worse, but the best results still come from the higher diff bin 10 days out (e.g. row 17) but the slopes are not consistently positive and most of the results are bad. 

## 3. Investigate leader/lagger relationship for index-future spread

_This code might need to be changed a bit since I belive it's still relying on the Bloomberg data._

Basically want to see if a certain spread between the future and index t days from expiration means that the future is likely to converge to the value of the index or vice versa.

First we'll compile the necessary data. This involves knowing how the future and index change in the last t days until expiration.

```python
# create dataframe that compares the difference between future & index t days out with
# the change in both the future and index over the remaining t days

# first, add changes to hog dataframe

for t in [2, 5, 10, 20]:
        # change in 1-day index for the last t days until expiry
        hog_df[f"Chg1day_{t}"] = abs(hog_df["1_Day_Indx"] - hog_df["1_Day_Indx"].shift(-t))
        # change in 2-day index for the last t days until expiry
        hog_df[f"Chg2day_{t}"] = abs(hog_df["2_Day_Indx"] - hog_df["2_Day_Indx"].shift(-t))
      
    
hog_df
```

Creating the dataframe with spreads and index future changes for a range of t.

```python
# this is basically going to be the same as the regression dataframe

change_df = pd.DataFrame()
change_df["Contract"] = contract_dict.keys()

# change_df = change_df.loc[(change_df["Contract"] != "DEC 19") & (change_df["Contract"] != "OCT 18") & (change_df["Contract"] != "APR 18")]

# something happens in sept-oct 2013 :(
change_df = change_df.loc[(change_df["Contract"] != "OCT 13")]

# drop latest contract?
change_df.drop(change_df.index[0], inplace=True)

for t in [20, 10, 5, 2]:
    # should be using t or t-1? is the last day 1 or 0 days to expiry? currently treating the last day as 0 days to expiry
    # absolute difference in future & 2 day rolling index t days from expiry
    change_df[f"Diff_{t}"] = change_df["Contract"].apply(lambda x: abs(contract_dict[x].loc[contract_dict[x].index[t],"Settle"] - hog_df.loc[contract_dict[x].index[t],"2_Day_Indx"]))
    
#     # absolute difference in future & 1 day implied index t days from expiry
#     change_df[f"Diff1_{t}"] = change_df["Contract"].apply(lambda x: abs(contract_dict[x].loc[contract_dict[x].index[t],"Settle"] - hog_df.loc[contract_dict[x].index[t],"1_Day_Implied"]))
   
    # absolute change in future price over the last t days of the contract
    change_df[f"Future_Chg_{t}"] = change_df["Contract"].apply(lambda x: contract_dict[x].loc[contract_dict[x].index[0],f"Change_{t}"])
 
    # absolute change in 2 day index price over the last t days of the contract
    # had this be the 20 days ending the day after expiry since that's what the value is
    change_df[f"Index_Chg_{t}"] = change_df["Contract"].apply(lambda x: hog_df.loc[hog_df.index[hog_df.index.get_loc(contract_dict[x].index[0])-1],f"Chg2day_{t}"])
    
#     # absolute change in 2 day index price over the last t days of the contract
#     # had this be the 20 days ending the day after expiry since that's what the value is
#     change_df[f"Index2_Chg_{t}"] = change_df["Contract"].apply(lambda x: hog_df.loc[hog_df.index[hog_df.index.get_loc(contract_dict[x].index[0])-1],f"Chg2day_{t}"])

#     # absolute change in 1 day index price over the last t days of the contract
#     change_df[f"Index1_Chg_{t}"] = change_df["Contract"].apply(lambda x: hog_df.loc[contract_dict[x].index[0],f"Chg1day_{t}"])


# the diff isn't quite equal to the sum/diff of changes because it's the diff t days out but the index 
# change is like from -1 days out to t-1 days out. so it's missing a day of change

change_df
```
